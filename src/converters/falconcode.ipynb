```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import matplotlib.pyplot as plt
import os, sys

sys.path.insert(0, "../")

from spec.enums import MainTableColumns as Cols, EventType
```

```python
base_dir = os.path.join("..", "..", "sample_data", "falconcode")
sqlite_path = os.path.join(base_dir, "raw", "falconcode.sqlite")
sqlalchemy_url = "sqlite:///" + sqlite_path
assert os.path.exists(sqlite_path), "SQLite file does not exist at path: " + sqlite_path
```

```python
sqlite_output = os.path.join(base_dir, "falconcode.sqlite")
output_url = "sqlite:///" + sqlite_output
```

```python
runs = pd.read_sql_table(
    "runs",
    sqlalchemy_url,
)
```

```python
runs
```

```python
problems = pd.read_sql_table(
    "problems",
    sqlalchemy_url,
)
problems
```

```python
column_mapping = {
    "id": Cols.ProblemID,
    "course_id": Cols.TermID,
}

term_mapping = {
    2: "Spring 2021",
    3: "Fall 2021",
    4: "Spring 2022",
    5: "Test Data", # Just 47 records; likely not real data
}

link_problems = problems.rename(columns=column_mapping)
link_problems[Cols.TermID] = link_problems[Cols.TermID].map(term_mapping)

link_problems.head()
```

```python
link_problems.to_sql("LinkProblem", output_url, if_exists="replace", index=False)
```

```python
import numpy as np

column_mapping = {
    "id": Cols.EventID,
    "student_id": Cols.SubjectID,
    "course_id": Cols.TermID,
    "problem_id": Cols.ProblemID,
    "timestamp": Cols.ServerTimestamp,
    "score": Cols.Score,
    "code_hash": Cols.CodeStateID,
}

main_table = runs.rename(columns=column_mapping)
main_table = main_table[main_table[Cols.TermID] != 5] # Drop test data from the last term
main_table[Cols.TermID] = main_table[Cols.TermID].map(term_mapping)
main_table[Cols.EventType] = np.where(main_table[Cols.Score] == -1, EventType.RunProgram, EventType.Submit)
main_table.loc[main_table[Cols.Score] == -1, Cols.Score] = None
main_table[Cols.Score] = main_table[Cols.Score] / 100

main_table
```

```python
main_table.to_sql("MainTable", output_url, if_exists="replace", index=False)
```
